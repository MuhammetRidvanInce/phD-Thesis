```python
"""TÜİK'te bulunan 2009 - 2020 yılları arası kurumsal sektör hesaplarındaki 'Kurumsal Sektör Hesapları (yıl)'
ismindeki excel dosyalarını düzenleyerek 'KHS_(yıl)' isimli excel dosyaları olarak kaydetmektedir. """ 
```

```python
import openpyxl as xl
import numpy as np
import pandas as pd
```

```python
names = ["2009", "2010", "2011", "2012","2013",
         "2014", "2015", "2016", "2017","2018",
         "2019", "2020"]
```

```python
for name in names:
    wb = xl.load_workbook("Kurumsal Sektör Hesapları" + " " + name +".xlsx")
    sheet = wb["t_"+ name]
    
    wb1 = "Kurumsal Sektör Hesapları" + " " + name +".xlsx"
    sheet1 = "t_"+ name
    
    print(wb1, "-----", sheet1)
    
    cols = ["Mali Olmayan Şİrketler", "Mali Şirketler", 
                             "Genel Devlet", "Hanehalkı-Hizmet", 
                             "Toplam Ekonomi", "Dış Alem", 
                             "Sektörler Toplamı"]

    df = pd.DataFrame(columns = cols + cols)
    source_use_cels = ["J", "K", "L", "M", "N", "O", "P","G", "F", "E", "D", "C", "B", "A"]
    rng = [range(6,16),range(17,26),range(30,46),range(47,59),
       range(63,69),range(70,79),range(82,88),range(89,97),]

    for x in rng:

        for r in x:
            liste = []

            row_value = sheet["I" + str(r)].value
            if row_value in df.index:
                row_value = row_value + "_" + str(r)

            for c in source_use_cels:
                value = sheet[c + str(r)].value
                liste.append(value)

            df.loc[row_value] = liste
            
    df.to_excel("KSH_" + name + ".xlsx")
    
```
